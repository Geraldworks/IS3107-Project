```python
import requests

# Necessary Headers
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept-Language": "en-US,en;q=0.9",
}

# Session to persist cookies
session = requests.Session()
session.headers = headers

# First try
base_url = "https://www.uniqlo.com/sg/api/commerce/v3/en/products?path=%2C%2C61590&limit=40&offset=0"

# First scrape try
main_page = session.get(base_url)
```

## Routes for specific information

#### Landing Page of Each Sub Category

base_url = "https://www.uniqlo.com/sg/api/commerce/v3/en/products?path=%2C%2C61590&limit=<total_product_number>&offset=0"       

check status         

All Items: response.get('result').get('items')     
Overall Rating: response.get('result').get('items').get('rating').get('average')     
Written Review Numbers: response.get('result').get('items').get('rating').get('count')      
Written Review Distribution: response.get('result').get('items').get('rating').get('rateCount')      
ProductID : response.get('result').get('items').get('productId')        
Price: response.get('result').get('items').get('prices'),get('base').get('value') (need to check if item is on promotion)       

#### Landing Page of Each product in sub category for reviews

base_url = "https://www.uniqlo.com/sg/api/commerce/v3/en/l1s/<product_id>/reviews?offset=0&limit=<review_numbers>"          

All Written Reviews: response.get('result').get('reviews')        

```python
all_products = main_page.json().get('result').get('items')

one_product_price = all_products[0].get('prices').get('base').get('value')
one_product_review_numbers = all_products[0].get('rating').get('count')
one_product_id = all_products[0].get('productId')
```

```python
print(one_product_price)
print(one_product_review_numbers)
print(one_product_id)
```

```python
review_url = f"https://www.uniqlo.com/sg/api/commerce/v3/en/l1s/{one_product_id.split('-')[0][1:]}/reviews?offset=0&limit={one_product_review_numbers}"
review_url
```

```python
response_reviews = session.get(review_url).json()
```
